```python
import os
import json
import csv

def convert_json_to_mot(input_json_path, output_txt_path):
    with open(input_json_path, "r") as f:
        data = json.load(f)

    results = []

    for frame_str, frame_data in data.items():
        frame_num = int(frame_str)
        tracks = frame_data.get("tracks_target_conf_bbox")

        # Skip if tracks is None or not a dict
        if not isinstance(tracks, dict):
            continue

        for track_id, track_data in tracks.items():
            bbox = track_data[2]
            conf = track_data[1] if isinstance(track_data[1], (float, int)) else -1

            if bbox and all(isinstance(val, int) for val in bbox):
                x1, y1, x2, y2 = bbox
                width = x2 - x1
                height = y2 - y1
                result_line = [
                    frame_num,
                    int(track_id),
                    x1,
                    y1,
                    width,
                    height,
                    round(conf, 6) if conf != -1 else -1,
                    -1, -1, -1
                ]
                results.append(result_line)

    results.sort(key=lambda x: (x[0], x[1]))

    with open(output_txt_path, "w", newline='') as f:
        writer = csv.writer(f)
        writer.writerows(results)

    print(f" Converted: {input_json_path} -> {output_txt_path}")


# Set your root input folder and output folder here
input_root = "/mnt/DATA/Abhin/DRDO_Project/Methods/OCL/OCL-bytetracker-original/OCLReID/output-result-ocl-bytetracker/DRDO-Day"
output_root = "/mnt/DATA/Abhin/DRDO_Project/Methods/eval/TrackEval/data/trackers/mot_challenge/MOT20-train/MPNTrack/data"

# Ensure the output root exists
os.makedirs(output_root, exist_ok=True)

# Traverse all subdirectories of the input folder
for dirpath, _, filenames in os.walk(input_root):
    for filename in filenames:
        if filename.endswith("_ocl_bytetracker_result.json"):
            input_file = os.path.join(dirpath, filename)

            # Extract track or video folder name (e.g., fi_track02)
            relative_path = os.path.relpath(dirpath, input_root)
            track_name = os.path.basename(relative_path)

            # Create output filename like: fi_track02.txt
            output_file = os.path.join(output_root, f"{track_name}.txt")

            convert_json_to_mot(input_file, output_file)
```

```python
import os
import shutil

def copy_mot_metadata(input_root, output_root):
    for dirpath, dirnames, filenames in os.walk(input_root):
        if "seqinfo.ini" in filenames:
            rel_path = os.path.relpath(dirpath, input_root)  # preserve relative structure
            out_folder = os.path.join(output_root, rel_path)
            os.makedirs(out_folder, exist_ok=True)

            # Copy seqinfo.ini
            src_seqinfo = os.path.join(dirpath, "seqinfo.ini")
            dst_seqinfo = os.path.join(out_folder, "seqinfo.ini")
            shutil.copy2(src_seqinfo, dst_seqinfo)
            print(f" Copied: {src_seqinfo} -> {dst_seqinfo}")

            # Copy gt/gt.txt if it exists
            gt_src = os.path.join(dirpath, "gt", "gt.txt")
            if os.path.exists(gt_src):
                gt_dst_folder = os.path.join(out_folder, "gt")
                os.makedirs(gt_dst_folder, exist_ok=True)
                gt_dst = os.path.join(gt_dst_folder, "gt.txt")
                shutil.copy2(gt_src, gt_dst)
                print(f" Copied: {gt_src} -> {gt_dst}")
            else:
                print(f"  Missing gt.txt in: {dirpath}")

# Example usage:
input_root = "/mnt/DATA/Abhin/DRDO_Project/Methods/boxmot-files/boxmot/assets/DRDO-Day/Test"
output_root = "/mnt/DATA/Abhin/DRDO_Project/Methods/eval/TrackEval/data/gt/mot_challenge/MOT20-train"

copy_mot_metadata(input_root, output_root)
```

```python
import os
import configparser

def fix_seqinfo_names(root_dir):
    for dirpath, _, filenames in os.walk(root_dir):
        if "seqinfo.ini" in filenames:
            seqinfo_path = os.path.join(dirpath, "seqinfo.ini")
            folder_name = os.path.basename(dirpath)

            config = configparser.ConfigParser()
            config.read(seqinfo_path)

            if "Sequence" not in config:
                print(f" No [Sequence] section in: {seqinfo_path}")
                continue

            old_name = config["Sequence"].get("name", "")
            config["Sequence"]["name"] = folder_name

            with open(seqinfo_path, "w") as f:
                config.write(f)

            print(f" Updated 'name' in {seqinfo_path}: '{old_name}'  '{folder_name}'")

# Example usage:
root_dir = "/mnt/DATA/Abhin/DRDO_Project/Methods/boxmot-files/boxmot/assets/DRDO-Night/Test"
fix_seqinfo_names(root_dir)
```
